Criar um programna que leia diois numeros floats e apresente o resultado das operações:
Adição, Subtração, Multiplicação, divisão e Potenciação.

```python
#Entrada de dados
a = float (input("Infome o 1° número: "))
b = float (input("Infome o 2° número: "))
#Processamento de Dados
s=a+b
sub=a-b
m=a*b
d=a/b
p=a**b#calculo da potenciação - 'a' elevado a 'b'

#Saida dos resultados
print(f"{a} + {b} = {s}")
print(f"{a} - {b} = {sub}")
print(f"{a} * {b} = {m}")
print(f"{a} / {b} = {d}")
print(f"{a} ** {b} = {p}")
```
