```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_json('netflix_titles.json')

traducao = {
	"show_id":"id",
    "type":"tipo",
    "title":"titulo",
    "director":"diretor",
    "cast":"elenco",
    "country":"pais",
    "date_added":"data_adicionada",
    "release_year":'ano_lancamento',
    "rating":"avaliacao",
    "duration":"duracao",
    "listed_in":"categoria",
    "description":"descricao"
}

df = df.rename(columns=traducao)

print(df.columns)
```

```python
df['diretor'] = df['diretor'].fillna('Não Informado')
```

```python
df['elenco'] = df['elenco'].fillna('Não Informado')
```

```python
top_10_paises = df['pais'].value_counts().head(10)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='tipo', hue='tipo', palette='coolwarm', legend=False)

plt.title('Distribuição do Catálogo: Filmes vs Séries', fontsize=14)
plt.xlabel('Tipo de Conteúdo (Filme ou Série)')
plt.ylabel('Quantidade Total')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Atribuímos o index (países) ao 'hue' e definimos legend=False
sns.barplot(
    x=top_10_paises.values, 
    y=top_10_paises.index, 
    hue=top_10_paises.index, 
    palette='Spectral', 
    legend=True
)
plt.title('Top 10 Países que mais produzem para a Netflix')
plt.xlabel('Total de Títulos (Filmes e Séries)')
plt.ylabel('País')
plt.show()
```

Pergunta: "A Netflix é uma plataforma puramente americana ou os dados mostram uma diversidade internacional relevante?"

Análise dos Resultados:
Ao analisarmos o gráfico dos 10 países com mais títulos no catálogo, chegamos às seguintes conclusões:  Liderança Americana: Os Estados Unidos ocupam o primeiro lugar de forma isolada, o que confirma a origem e a força da base da plataforma.

Diversidade Internacional Relevante: No entanto, os dados refutam a ideia de uma plataforma "puramente americana". A Índia aparece como o segundo maior produtor (impulsionada por sua massiva indústria cinematográfica), seguida por potências como o Reino Unido, Canadá e França.

Mercados em Ascensão: A presença de países como Coreia do Sul, Espanha e Turquia entre os maiores produtores demonstra uma estratégia clara de globalização de conteúdo, onde produções não-americanas ganham destaque e alcançam audiências mundiais. 

Conclusão: Embora a influência dos EUA seja predominante, a Netflix evoluiu para um ecossistema global. A diversidade internacional é um pilar estratégico, com mais de 40% dos títulos do Top 10 vindo de países fora do eixo América do Norte/Reino Unido, provando que a plataforma é um palco para histórias de diversas culturas.
